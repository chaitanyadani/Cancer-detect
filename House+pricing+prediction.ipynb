```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
labels = h_data['price']
conv_dates = [1 if values == 2014 else 0 for values in h_data.date]
h_data['date'] = conv_dates
train1 = h_data.drop(['id', 'price'], axis = 1)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(train1, labels, test_size = 0.10, random_state = 2)
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_test, y_test)
```

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2, learning_rate = 0.2, loss = 'ls')
```

```python
clf.fit(x_train,y_train)
```
